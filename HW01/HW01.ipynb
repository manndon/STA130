The difference between an "attribute" and a "method" is that an "attribute" is a property of an object, that holds information relating to the object and that can be easily retrieved. For example, in df.shape, "shape" is an "attribute" of the df object that holds information about the amount of rows and columns in df. On the other hand, a "method" is a function (set of instructions) that is called within the context of an object. "Methods" have access to the "attributes" of the object that is calling it, and they can modify the object itself.

**'50%':** This is the 50th percentile value, also known as the median. It's the value that is larger than or equal to 50% of the values in the column.<br>
**'75%':** This is the 75th percentile value. It's the value that is larger than or equal to 75% of the values in the column.<br>
**'max' :** This is the largest value in the column.

The values in count that you get from doing `df.groupby("col1")["col2"].describe()` are fundamentally different from the values in count you get from `df.describe()`, because while the count values you obtain by doing `df.describe()` only report the amount of non-missing values for each column, the count values you obtain by doing `df.groupby("col1")["col2"].describe()` tell you how many observations are in each group; in other words, for all the different values in the "col1" column, it'll tell you how many observations have that specific value for the "col1" variable.
