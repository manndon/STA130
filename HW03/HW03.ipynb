````python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# Load the dataset
penguins = pd.read_csv('penguins.csv')

# Function to add statistics
def add_stat_lines_and_rectangles(fig, species, data, color, row, col):
    mean_val = data.mean()
    median_val = data.median()
    min_val = data.min()
    max_val = data.max()
    std_dev = data.std()
    
    # Interquartile range (IQR)
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    
    # Add shaded area for Range
    fig.add_vrect(x0=min_val, x1=max_val, fillcolor=color, opacity=0.1, line_width=0, row=row, col=col)
    
    # Add shaded area for Interquartile Range (IQR)
    fig.add_vrect(x0=q1, x1=q3, fillcolor=color, opacity=0.2, line_width=0, row=row, col=col)
    
    # Add shaded area for two standard deviations away from the mean
    fig.add_vrect(x0=mean_val - 2 * std_dev, x1=mean_val + 2 * std_dev, fillcolor=color, opacity=0.3, line_width=0, row=row, col=col)
    
    # Add vertical lines for mean and median after shaded areas (to be in the foreground)
    fig.add_vline(x=mean_val, line=dict(color='red', dash='solid', width=4), row=row, col=col)
    fig.add_vline(x=median_val, line=dict(color='purple', dash='dash', width=4), row=row, col=col)

# Create subplots: One row, three columns for each species
fig = make_subplots(rows=1, cols=3, subplot_titles=penguins['species'].unique())

# Add histograms and statistics for each species
for i, (species, color) in enumerate(zip(penguins['species'].unique(), ['blue', 'green', 'orange'])):
    species_data = penguins[penguins['species'] == species]['flipper_length_mm']
    
    # Add histogram to the corresponding subplot
    fig.add_trace(go.Histogram(x=species_data, name=f'{species} flipper length', marker_color=color, opacity=0.75, showlegend=False), row=1, col=i+1)
    
    # Add statistical lines and rectangles
    add_stat_lines_and_rectangles(fig, species, species_data, color, row=1, col=i+1)

# Add a legend for the statistical markers
# Using a dummy trace for legend purposes
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', line=dict(color='red', width=4), name='Mean (solid red)'))
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', line=dict(color='purple', width=4, dash='dash'), name='Median (dashed purple)'))
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', line=dict(color='black', width=1), name='Range (shaded area)'))
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', line=dict(color='black', width=1), name='IQR (darker shaded area)'))
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', line=dict(color='black', width=1), name='2 Std Dev (lightest shaded area)'))

# Update layout
fig.update_layout(title='Flipper Length Distributions by Species with Statistical Markers',
                  xaxis_title='Flipper Length (mm)',
                  yaxis_title='Count',
                  height=600, width=1200)

# Show the plot
fig.show(renderer="png")
````

````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the penguins dataset
penguins = pd.read_csv("penguins.csv")

# Function to add statistics to each subplot
def add_stat_lines_and_rectangles(ax, data, color):
    mean_val = data.mean()
    median_val = data.median()
    min_val = data.min()
    max_val = data.max()
    std_dev = data.std()
    
    # Interquartile range (IQR)
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    
    # Add shaded area for Range
    ax.axvspan(min_val, max_val, color=color, alpha=0.1, label="Range")
    
    # Add shaded area for Interquartile Range (IQR)
    ax.axvspan(q1, q3, color=color, alpha=0.2, label="IQR")
    
    # Add shaded area for two standard deviations away from the mean
    ax.axvspan(mean_val - 2 * std_dev, mean_val + 2 * std_dev, color=color, alpha=0.3, label="2 Std Dev")
    
    # Add vertical lines for mean and median
    ax.axvline(mean_val, color='red', linestyle='-', linewidth=3, label='Mean (red)')
    ax.axvline(median_val, color='purple', linestyle='--', linewidth=3, label='Median (purple)')

# Set up the figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Define colors for each species' KDE plot and fill
colors = ['blue', 'green', 'orange']

# Loop over each species to create a KDE plot
for i, (species, color) in enumerate(zip(penguins['species'].unique(), colors)):
    species_data = penguins[penguins['species'] == species]['flipper_length_mm']
    
    # Plot KDE for the species
    sns.kdeplot(species_data, ax=axes[i], color=color, fill=True, alpha=0.6)
    
    # Add statistical markers to the plot
    add_stat_lines_and_rectangles(axes[i], species_data, color)
    
    # Set plot title and labels
    axes[i].set_title(f'{species} Flipper Length Distribution')
    axes[i].set_xlabel('Flipper Length (mm)')
    axes[i].set_ylabel('Density')

# Add a shared legend for all plots
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=5)

# Adjust the layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Display the plot
plt.show()
````

**If the mean is larger than the median, there is right skew, whereas if the median is larger than the mean, there is left skew. The reason for this is that the mean is more sensitive to extreme (relatively very low or very high) values in a dataset than the median. Since the median stays with the bulk of the data due to its insensensitivy to extreme values, we can compare it to the mean to get an idea of which way (left or right) the outliers are. We can understand this better using examples:**

````python
from scipy import stats
import pandas as pd
import numpy as np
import plotly.express as px

sample1 = stats.gamma(a=2,scale=2).rvs(size=1000)
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data")
fig1.show(renderer="png")
````

**By looking at the histogram above, we can obviously see that the data has right skew, as the majority of the data points are on the left side of the histogram. Let's see what the median and mean values of the dataset are:**

````python
print(f"Mean: {sample1.mean()}")
print(f"Median: {np.quantile(sample1, [0.5])[0]}")
````

**As we can see, the mean is larger than the median, which is why we can confirm numerically that the dataset has right skew. Now let's see an example of a dataset with left skew:**

````python
sample2 = -stats.gamma(a=2,scale=2).rvs(size=1000)
fig2 = px.histogram(pd.DataFrame({'data': sample2}), x="data")
fig2.show(renderer="png")
````

**Again, we can graphically see that the data has left skew, but we can prove it numerically by comparing the mean and the median values of the data:**

````python
print(f"Mean: {sample2.mean()}")
print(f"Median: {np.quantile(sample2, [0.5])[0]}")
````

**As opposed to the last example, here, the median is larger than the mean, which is why we can confirm numerically that the dataset has left skew.**

**Next, I asked ChatGPT to plot the average ratings of movies grouped by the decade they were released in using Seaborn. This is the code it came up with:**

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure startYear is numeric
movies['startYear'] = pd.to_numeric(movies['startYear'], errors='coerce')

# Create a new column for the decade
movies['decade'] = (movies['startYear'] // 10) * 10

# Group by the decade and calculate the mean of the averageRating
decade_ratings = movies.groupby('decade')['averageRating'].mean().reset_index()

# Plot using Seaborn as a histogram-like bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='decade', y='averageRating', data=decade_ratings, color="b")

# Add labels and title
plt.xlabel('Decade')
plt.ylabel('Average Rating')
plt.title('Average Movie Ratings by Decade')

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.show()
````

**As we can see in the plot above, the average rating of movies has gone down almost every decade since 1920. The average rating of a movie released in the 1920s is a little over 8/10, while the average rating of a movie released during this century is a little under 7/10.**

**It is important to note that part of the reason the older movies in the dataset have a higher average rating is because I set a limit of 50 000 minimum ratings for each movie to be included in the dataset. So, it's possible that there are worse movies from this time that weren't included in the dataset because people don't bother to watch and rate them as much. However, we can still deduce that there is a general downward trend in the quality of movies over time.**

**I also asked ChatGPT to plot the average rating of movies by their running time. This is the code it came up with:**

````python
# Ensure runtimeMinutes is numeric
movies['runtimeMinutes'] = pd.to_numeric(movies['runtimeMinutes'], errors='coerce')

# Create bins for running time
bins = range(0, movies['runtimeMinutes'].max() + 20, 20)  # Create bins of width 20 minutes
movies['runtime_bins'] = pd.cut(movies['runtimeMinutes'], bins=bins)

# Group by the running time bins and calculate the mean of the averageRating
runtime_ratings = movies.groupby('runtime_bins')['averageRating'].mean().reset_index()

# Plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='runtime_bins', y='averageRating', data=runtime_ratings, color="b")

# Add labels and title
plt.xlabel('Runtime (Minutes)')
plt.ylabel('Average Rating')
plt.title('Average Movie Ratings by Running Time')

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.show()
````

**As we can see in the plot above, movies that last between 1 hour 20 minutes and 1 hour 40 minutes are generally worse than movies that are longer or shorter. By contrast, movies that are longer than 3 hours are generally rated higher than movies shorter than 3 hours. Lastly, movies that last between 40 minutes and an hour are also generally highly rated.**
